```python
import csv
import urllib.request

res = "http://opendata.hccg.gov.tw/dataset/432257df-491f-4875-8b56-dd814aee5d7b/resource/de014c8b-9b75-4152-9fc6-f0d499cefbe4/download/20150305140446074.csv"
urllib.request.urlretrieve(res, './data/example.csv')

```

```python
with open('./data/example.csv', newline='', encoding='utf-8') as csvfile:
    # 讀取 CSV 檔案內容
    rows = csv.reader(csvfile)
    for row in rows:
        time.append(row[5])
    print(time)
        
```

```python
data = {}
one = []
two = []
three = []
four = []
five = []
header = True

with open('./data/example.csv', newline='', encoding='utf-8') as csvfile:
    # 讀取 CSV 檔案內容
    rows = csv.reader(csvfile)
    for row in rows:
        if header:
            header = not header
            continue
        one.append(row[5])
        two.append(row[6])
        three.append(row[7])
        four.append(row[8])
        five.append(row[9])
        
        data['班次1'] = one
        data['班次2'] = two
        data['班次3'] = three
        data['班次4'] = four
        data['班次5'] = five
        
    print(data)    


```
